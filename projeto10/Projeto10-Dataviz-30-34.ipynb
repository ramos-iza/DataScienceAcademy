> Dataviz 30 - Gráfico de barras na horizontal com o tráfego aéreo por país

```python
df["Country"].unique()
```

```python
# Dataviz 30
sns.set(font_scale = 2)
plt.figure(figsize = (16,8))
sns.barplot(x = "PercentOfBaseline", y = "Country", data = df)
plt.title("Tráfego Aéreo Por País")
plt.xlabel('Tráfego Aéreo')
plt.ylabel('País')
```

> Dataviz 31 - Gráfico de pizza com o tráfego aéreo por país

```python
# Ajusta os dados
trafego_paises = df['Country'].value_counts().to_frame('Count')
trafego_paises
```

```python
# Dataviz 31
sns.set(font_scale = 1)
plt.figure(figsize = (8, 8))
plt.pie(trafego_paises['Count'], 
        labels = trafego_paises.index, 
        autopct = '%1.1f%%')
plt.show()
```

> Dataviz 32 - Gráfico de barras na horizontal com o tráfego aéreo por cidade

```python
df["City"].unique()
```

```python
# Dataviz 32
sns.set(font_scale = 2)
plt.figure(figsize = (20,18))
sns.barplot(x = df.PercentOfBaseline, y = df.City, capsize = .2)
plt.title("Tráfego Aéreo Por Cidade")
plt.xlabel('Tráfego Aéreo')
plt.ylabel('Cidade')
```

> Dataviz 33 - Gráfico de barras na horizontal com o tráfego aéreo por aeroporto

```python
df["AirportName"].unique()
```

```python
# Dataviz 33
sns.set(font_scale = 2)
plt.figure(figsize = (25,25))
sns.barplot(x = df.PercentOfBaseline, y = df.AirportName, palette = "Blues_d", capsize = .2)
plt.title("Tráfego Aéreo Por Aeroporto")
plt.xlabel('Tráfego Aéreo')
plt.ylabel('Aeroporto')
```

> Dataviz 34 - Scatter Plot com o tráfego aéreo por estado

```python
# Dataviz 34
sns.set(font_scale = 2)
plt.figure(figsize = (18,16))
sns.scatterplot(y = df.State, x = df.PercentOfBaseline, s = 100)
plt.xlabel('Tráfego Aéreo')
plt.ylabel('Estado')
```
