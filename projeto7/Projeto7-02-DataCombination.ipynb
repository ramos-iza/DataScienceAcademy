```python
# Imports
import json
import pandas as pd
```

> Primeiro carregamos e combinamos os dados de AnimeInfo.

```python
# Cria a lista
animeData = []
```

```python
# Loop para carregar os arquivos do disco na lista
for i in range(0, 451, 50):
    with open(f"/Users/izadoraramos/Desktop/DataScienceAcademy/projeto7/dados/AnimeInfo/animeInfo{i}.txt") as file:
        animeInfo = json.load(file)
        for j in range(50):
            animeData.append(animeInfo[j])
            
```

```python
# Converte a lista para dataframe
df1 = pd.DataFrame(animeData)
```

```python
# Visualiza amostra
df1.head()
```

```python
# Info
df1.info()
```

> Agora carregamos e combinamos os dados de animeInfoNames.

```python
# Cria a lista
animeNames = []
```

```python
# Loop para carregar os arquivos do disco na lista
for i in range(0, 451, 50):
    with open(f"/Users/izadoraramos/Desktop/DataScienceAcademy/projeto7/dados/AnimeInfoNames/animeInfoNames{i}.txt") as file:
        animeInfoNames = json.load(file)
        for j in range(50):
            animeNames.append(animeInfoNames[j])
```

```python
# Converte em dataframe
df2 = pd.DataFrame(animeNames)
```

```python
df2.head()
```

## Combinando os Dataframes

```python
# Cria uma cópia de df1
df3 = df1.copy()
```

```python
# Combina df3 e df2
df3["name"] = df2["name"]
```

```python
# Visualiza amostra
df3.head()
```

```python
# Salva em disco
df3.to_csv("/Users/izadoraramos/Desktop/DataScienceAcademy/projeto7/dados/Processamento/animeInfoCombination.csv", index = False)
```
